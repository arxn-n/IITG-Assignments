```python
#Question

'''

Apriori is a statistical algorithm for implementing associate rule mining,
that primarily relies on three components: Life, Support and Confidence. 
Using this algorithm try to find the rules that describe the relation between
each of the products that were brought by the customers

'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv')
```

```python
store_data.head()
```

```python
store_data = pd.read_csv('store_data.csv', header=None)
```

```python
store_data.head()
```

```python
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
print(association_results)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
results=[]
for item in association_results:
    pair = item[0] 
    items = [x for x in pair]

    value0=str(items[0])
    value1=str(items[1])
    value2=str(item[1])[:7]
    value3=str(item[2][0][2])[:7]
    value4=str(item[2][0][3])[:7]
    
    rows=(value0,value1,value2,value3,value4)
    
    results.append(rows)
    
    Label=['Title1','Title2','Support','Confidence','Lift']
    
    store_suggestion=pd.DataFrame.from_records(results, columns=Label)
```
